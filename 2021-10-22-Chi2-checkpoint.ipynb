```python
import pandas as pd
from scipy import stats
df_pay = pd.read_csv(r'C:\Users\anton\OneDrive\Desktop\Проги\Tests\payments.csv') 
df_play = pd.read_csv(r'C:\Users\anton\OneDrive\Desktop\Проги\Tests\sessions.csv') 
```

Сгруппируем таблицу с играми и присоеденим информацию о пользовтаелях к таблице с платежами

```python
df_test = df_play.groupby(['#1 user id'])['#2 Test group'].max()
df_test = df_test.reset_index()
df_test = df_pay.merge(df_test, how = 'left', on = '#1 user id')
df_test.head()
```

Как видно, далеко не все пользователи пришли из нашего А/Б теста.
Посчитаем тех пользователей, кто из А/Б теста дошел до триала или оплаты.

```python
test_trial = df_test.loc[(df_test['#2 Test group']=='05_12FTD_test_group') 
                         & (df_test['#3 is_trial']==True)
                         ]['#3 is_trial'].count()
test_pay = df_test.loc[(df_test['#2 Test group']=='05_12FTD_test_group') 
                         & (df_test['#3 is_trial']==False)
                         ]['price'].value_counts()
control_trial = df_test.loc[(df_test['#2 Test group']=='05_12FTD_control_group') 
                         & (df_test['#3 is_trial']==True)
                         ]['#3 is_trial'].count()
control_pay = df_test.loc[(df_test['#2 Test group']=='05_12FTD_control_group') 
                         & (df_test['#3 is_trial']==False)
                         ]['price'].value_counts()

print('Control group:', 
      '\nsubjects: ',df_play['#2 Test group'].value_counts()['05_12FTD_control_group'],
      '\nsubscribed: ',df_test['#2 Test group'].value_counts()['05_12FTD_control_group'],
      '\ntrial: ',control_trial,'\npaid:\n', control_pay      )

print('Test group:', 
      '\nsubjects: ',df_play['#2 Test group'].value_counts()['05_12FTD_test_group'],
      '\nsubscribed: ',df_test['#2 Test group'].value_counts()['05_12FTD_test_group'],
      '\ntrial: ',test_trial,'\npaid:\n', test_pay      )
```

Теперь воспользуемся тестом Хи-квадрат для определения статистической значимости

```python
A = df_play['#2 Test group'].value_counts()['05_12FTD_control_group']
a = df_test['#2 Test group'].value_counts()['05_12FTD_control_group']
B = df_play['#2 Test group'].value_counts()['05_12FTD_test_group']
b = df_test['#2 Test group'].value_counts()['05_12FTD_test_group']
print ('All A\B subjects:',stats.chi2_contingency(([a,A-a],[b,B-b]),correction=False)[:2])
print ('Trial:',stats.chi2_contingency(([control_trial,a-control_trial],[test_trial,b-test_trial]),correction=False)[:2])
print ('Simple subscription:',stats.chi2_contingency(([control_pay.iloc[0],a-control_pay.iloc[0]]
                                                      ,[test_pay.iloc[0],b-test_pay.iloc[0]])
                                                     ,correction=False)[:2])
print ('Complex subscription:',stats.chi2_contingency(([0,a]
                                                      ,[test_pay.iloc[1],b-test_pay.iloc[1]])
                                                     ,correction=False)[:2])
```

Как видно, конвертация из тестовой и контрольной группы в заинтересованных пользователей не имеет статистически 
значимых различий (p =0.65).
Статистических различий между конвертациями в разные виды подписки тоже не наблюдается:
Триал: p = 0.66, простая подписка: p = 0.40, комплексаня подписка: p = 0.07.
